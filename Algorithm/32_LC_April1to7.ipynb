### Happy Number (solved)
https://leetcode.com/explore/challenge/card/30-day-leetcoding-challenge/528/week-1/3284/

```python
def isHappy(n: int) -> bool:
    def one_to_two(k):
        return int(str(k) + '0')
    
    if (len(str(n)) == 1) and (n == 1):
        return True

    else:
        square_number = n**2
        loop_check = 0
        if len(str(n)) == 1:
            n = one_to_two(n)
        
        while (len(str(n)) > 1):
            res = 0
            for i in list(str(n)):
                res += int(i) ** 2
                loop_check += 1

                if (loop_check >= 2) and (res == square_number):
                    return False

            n = res
            if res == 89:
                return False
            
            if (len(str(n)) == 1) and (n != 1):
                n = one_to_two(n)
        
        return True if n == 1 else False
```

### Counting Elements (solved)
https://leetcode.com/explore/featured/card/30-day-leetcoding-challenge/528/week-1/3289/

```python
def countElements(arr: List[int]) -> int:
    assert 1 <= len(arr) <= 1000
    assert (max(arr) <= 1000) and (min(arr) >= 0)
    
    from collections import Counter
    cnt_arr = Counter(arr)
    cnt = 0
    for i in cnt_arr.keys():
        if i+1 in cnt_arr.keys():
            cnt += cnt_arr[i]
            
        else:
            continue
            
    print(cnt)
    return cnt
```

```python
a = [1,2,3]
print('a:',countElements(a))
print('-'*50)
b = [1,1,3,3,5,5,7,7]
print('b:',countElements(b))
print('-'*50)
c = [1,3,2,3,5,0]
print('c:',countElements(c))
print('-'*50)
d = [1,1,2,2]
print('d:',countElements(d))
print('-'*50)
```
