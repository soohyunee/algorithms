### Running sum of 1D-array (05/03)
https://leetcode.com/problems/running-sum-of-1d-array/

```python
## Runtime: 36ms / faster than 84.39%
## Memory usage: 14.4mb / less than 43.79%
def runningSum(nums: List[int]) -> List[int]:
    for idx in range(1,len(nums)):
        nums[idx] += nums[idx-1]
    return nums
```

```python
a = [1,2,3,4]
runningSum(a)
```

```python
b = [1,1,1,1,1]
runningSum(b)
```

```python
c = [3,1,2,10,1]
runningSum(c)
```

### Array Partition (05/07)
https://leetcode.com/problems/array-partition-i/

```python
## Runtime: 280ms / faster than 25.07%
## Memory usage: 16.8mb / 60.24%
import time
def arrayPairSum(nums: List[int]) -> int:
    start = time.time()
    nums.sort()
    res = 0
    idx = 0
    while idx < len(nums)-1:
        res += min(nums[idx],nums[idx+1])
        idx += 2
    print(time.time() - start)
    return res
```

### Evaluate Reverse Polish Notation (05/25)
https://leetcode.com/problems/evaluate-reverse-polish-notation/<br>
1. operator 발견시에만 앞에 있는 숫자들(정방향)을 체크해서 가져옴 (stack, queue 둘 다 필요)<br>
연산할때는 stack, 숫자를 불러오는 건 queue 형태여야 함.. (순서가 필요하니깐)<br>
2. 한 번 연산이 이뤄지면 그 값에서 변화가 있어야함<br>

```python
## Runtime: 56ms / faster than 98.09%
## Memory usage: 14.7mb / 42.12%
# reference: https://leetcode.com/problems/evaluate-reverse-polish-notation/discuss/1229359/Python-short-stack-solution-explained
## 보면 stack과 queue의 성질을 모두 필요로 하는 문제인데,
## stack에 결과물을 다시 집어넣음으로써, stack 하나만으로 풀었음 @@;;
## 다만 빼기와 나눗셈 연산은, 순서가 다른데.. 정확한 이유를 모르겠음. 내가 코드짤때도 역으로 하니
## 오히려 정답이 나왔기는 한데...
from typing import List
def evalRPN(tokens: List[str]) -> int:
    if len(tokens) <= 1:
        return int(tokens[0])

    stack = []
    
    def operation(val1, val2, operator):
        if operator == "+":
            return val1 + val2
        elif operator == "-":
            return val2 - val1
        elif operator == "*":
            return val1 * val2
        else:          # "/"
            return int(val2 / val1)
    
    for idx in range(len(tokens)):
        if tokens[idx] in ["+","-","*","/"]:
            stack.append(operation(stack.pop(), stack.pop(), tokens[idx]))
        else:
            stack.append(int(tokens[idx]))
            
    return stack[-1]
```

```python
a = ["2","1","+","3","*"]
evalRPN(a)   # 9
```

```python
b = ["4","13","5","/","+"]
evalRPN(b)   # 6
```

```python
c = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
evalRPN(c)   # 22
```

```python
d = ["18"]
evalRPN(d)     # 18
```

```python
e = ["3","11","+","5","-"]
evalRPN(e)       # 9
```

### Maximum Product of Word Lengths (05/28)
https://leetcode.com/problems/maximum-product-of-word-lengths/<br>
1. 곱해지는 두 단어간에는 공통된 letter가 없어야 함<br>
2. frozenset(built-in function)을 사용해서 풀 수도 있음<br>

```python
## Runtime: 948ms / faster than 52.55%
## Memory usage: 16.1mb / less than 5.42%
from typing import List
def maxProduct(words: List[str]) -> int:
    word_set = [set(w) for w in words]
    max_ret = 0
    length = len(word_set)
    for idx1 in range(length):
        for idx2 in range(idx1+1, length):
            # no common element between idx1 and idx2
            if len(word_set[idx1] - word_set[idx2]) == len(word_set[idx1]):
                if len(words[idx1]) * len(words[idx2]) > max_ret:
                    max_ret = len(words[idx1]) * len(words[idx2])
                
    return max_ret
```

```python
a = ["abcw","baz","foo","bar","xtfn","abcdef"]
maxProduct(a)   # 16
```

```python
b = ["a","ab","abc","d","cd","bcd","abcd"]
maxProduct(b)   # 4
```

```python
c = ["a","aa","aaa","aaaa"]
maxProduct(c)        # 0
```
