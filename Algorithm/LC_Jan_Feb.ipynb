### Count Largest Group (01/02) 
https://leetcode.com/problems/count-largest-group/<br>
1. 10 이후부터의 숫자는 그 summation의 위치에 맞게 들어가야 함<br>

```python
## These codes are too inefficient, let's make them more efficiently..

## Runtime: 192ms / faster than 5.28%
## Memory usage: 14.4mb / less than 22.94%
def countLargestGroup(n: int) -> int:
    tmp_list, tmp_res = [], 0
    ranges = list(range(1,n+1))
    track = []
    
    while ranges:
        value = ranges.pop(0)
        
        if value < 10:
            track.append(value)
            tmp_list.append([value])
    
        else:
            for x in str(value):
                tmp_res += int(x)
                
            if tmp_res not in track:
                track.append(tmp_res)
                tmp_list.append([value])
                tmp_res = 0
                
            elif tmp_res in track:
                tmp_list[tmp_res-1].append(value)
                tmp_res = 0
                
       
    max_len, cnt = 1, 0
    for i in tmp_list:
        if len(i) > max_len:
            max_len = len(i)

    for j in tmp_list:
        if len(j) == max_len:
            cnt += 1

    return cnt
```

```python
countLargestGroup(n=13)  # 4
```

```python
countLargestGroup(n=2)  # 2
```

```python
countLargestGroup(n=15)  # 6
```

```python
countLargestGroup(n=24)  # 5
```

```python
countLargestGroup(n=264)  # 2
```

### K-th Missing Positive Number (1/6)
https://leetcode.com/problems/kth-missing-positive-number/

```python
## Runtime: 80ms / faster than 26.20%
## Memory usage: 14.3mb / less than 63.22%
def findKthPositive(arr: List[int], k: int) -> int:
    max_num = max(arr)
    whole_arr = list(range(1,max_num+1))
    
    for i in arr:
        if i in whole_arr:
            whole_arr.pop(whole_arr.index(i))
        
    if k < len(whole_arr):
        return whole_arr[k-1]
    
    elif k == len(whole_arr):
        return whole_arr[0]

    else: # k > len(whole_arr)
        for x in range(max_num+1,max_num+k+1):
            whole_arr.append(x)
        return whole_arr[k-1]
```

```python
a = [2,3,4,7,11]
findKthPositive(a,5)  # 9
```

```python
b = [1,2,3,4]
findKthPositive(b,2)  # 6
```

```python
c = [2]
findKthPositive(c,1)  # 1
```

## N-Repeated Element in Size 2N Array (1/30)
https://leetcode.com/problems/n-repeated-element-in-size-2n-array/

```python
## Runtime: 208 ms / faster than 52.75%
## Memory usage: 15.5mb / less than 72.87%
from typing import List
def repeatedNTimes(A: List[int]) -> int:
    n = len(A)/2
    dic = dict()
    for i in A:
        if i in dic.keys():
            dic[i] += 1
        else:
            dic[i] = 1

    for k,v in dic.items():
        if v == n:      # 여기서 n대신에 len(A)/2를 넣어주면, memory usage가 92% 정도로 올라감
            return k
```

```python
a = [1,2,3,3]
repeatedNTimes(a)
```

```python
b = [2,1,2,5,3,2]
repeatedNTimes(b)
```

```python
c = [5,1,5,2,5,3,5,4]
repeatedNTimes(c)
```

### Number of 1 Bits (02/07)
https://leetcode.com/problems/number-of-1-bits/

```python
## Runtime: 48ms / faster than 20.49%
## Memory usage: 14.3mb / less than 43.47%
def hammingWeight(n: int) -> int:
    cnt = 0
    n = str(bin(n))
    for i in n[2:]:
        if int(i) >= 1:
            cnt += 1

    return cnt
```

### Search in a Binary Search Tree (2/14 시도, 못품)
https://leetcode.com/problems/search-in-a-binary-search-tree/

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

## reference code: https://leetcode.com/problems/search-in-a-binary-search-tree/discuss/1061426/My-simple-and-fast-solution-in-Python
class Solution:
    def searchBST(self, root: TreeNode, val: int) -> TreeNode:
        if root is None:
            return None
        
        if root.val == val:
            return root
        
        elif root.val < val:
            return self.searchBST(root.right, val)
            
        elif root.val > val:
            return self.searchBST(root.left, val)
```

### Find the Highest Altitude (2/14)
https://leetcode.com/problems/find-the-highest-altitude/

```python
## Runtime: 24ms / faster than 99.35%     O(N)
## Memory Usage: 14.2mb / less than 44.49%
def largestAltitude(gain: List[int]) -> int:
    res = [0,]
    tmp = 0
    max_value = 0
    for i in gain:
        tmp += i
        res.append(tmp)
        if tmp > max_value:
            max_value = tmp
    return max_value
```

```python
a = [-5,1,5,0,-7]
largestAltitude(a)
```

```python
b = [-4,-3,-2,-1,4,3,2]
largestAltitude(b)
```
